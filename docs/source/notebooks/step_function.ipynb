```python
from hbtpy.soicc import calculate_g2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

```python
sf = np.ones(90)
sf[60:75] = -1
sf[5:10] = -1
 
#sf[10:25] = 2
# sf = np.array([sf for a in range(10)])
#  sf
 
t = np.linspace(0,90,90)
```

```python
corr = np.zeros([sf.shape[0], sf.shape[0]])

for i in range(sf.shape[0]):
    for j in range(sf.shape[0]):
        corr[-i,-j] = (sf[i]*sf[j])/(sf[i]**2)
```

```python

fig, ax = plt.subplots(1,1, figsize = (8,7))


im = ax.imshow(np.fliplr(corr),
          extent = [0,90,0,90],
          cmap = 'turbo')

ax.set_aspect('auto')

ax.set_xlabel("$t_1$ ($\mu$s)")
ax.set_ylabel("$t_2$ ($\mu$s)")

ax.set_xticks([0,30,60,90])
ax.set_yticks([0,30,60,90])

cbar = fig.colorbar(im)
cbar.set_ticks([-1,0,1])
cbar.set_label("Correlation, $g^{(2)}$", fontsize = 22)
matplotlib.rcParams.update({'font.size': 22})
ax.tick_params(axis='both', which='both', labelsize=22)

fig.savefig("/home/guestt/Pictures/corr_demo1.png")
```
